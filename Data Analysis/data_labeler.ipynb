```python
import pandas as pd

# Load the CSV file
file_path = 'walk up and down staircase mixed.csv'  # Change this to your actual file path
df = pd.read_csv(file_path)

```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Create a figure with subplots
fig = make_subplots(rows=1, cols=2, subplot_titles=("Acceleration Data", "Gyroscope Data"))

# Acceleration Data
fig.add_trace(go.Scatter(x=df.index, y=df['acceleration_x'], mode='lines+markers', name='Accel X'), row=1, col=1)
fig.add_trace(go.Scatter(x=df.index, y=df['acceleration_y'], mode='lines+markers', name='Accel Y'), row=1, col=1)
fig.add_trace(go.Scatter(x=df.index, y=df['acceleration_z'], mode='lines+markers', name='Accel Z'), row=1, col=1)

# Gyroscope Data
fig.add_trace(go.Scatter(x=df.index, y=df['gyroscope_x'], mode='lines+markers', name='Gyro X'), row=1, col=2)
fig.add_trace(go.Scatter(x=df.index, y=df['gyroscope_y'], mode='lines+markers', name='Gyro Y'), row=1, col=2)
fig.add_trace(go.Scatter(x=df.index, y=df['gyroscope_z'], mode='lines+markers', name='Gyro Z'), row=1, col=2)

# Update xaxis properties
fig.update_xaxes(title_text="Measurement Index", row=1, col=1)
fig.update_xaxes(title_text="Measurement Index", row=1, col=2)

# Update yaxis properties
fig.update_yaxes(title_text="Acceleration (g)", row=1, col=1)
fig.update_yaxes(title_text="Angular Velocity (degrees/sec)", row=1, col=2)

# Update layout for the figure
fig.update_layout(height=600, width=1200, title_text="Sensor Data Visualization", showlegend=True)

# Show the figure
fig.show()
```
